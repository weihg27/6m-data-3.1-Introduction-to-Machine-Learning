## 🎯 Category 1 — Supervised Learning

**The idea in plain English.** You have inputs *and* known correct answers. The model learns to map inputs to answers so it can predict answers for new, unseen inputs.

**Real-world analogy.** A student studying past exam papers *with answer keys*. They see many questions and the correct answers, and over time learn to answer new questions on their own.

**Sarah's sentiment task is supervised.** Before she loaded the pre-trained DistilBERT model, someone had labelled millions of sentences as positive or negative. The model trained on those, learned the patterns, and is now used on *Sarah's* 10,000 unlabelled reviews. She does the last step (inference) — they did the first step (training).

**Why it matters.** Supervised learning is **the workhorse of business ML**. Churn prediction, fraud detection, demand forecasting, sentiment analysis — all supervised. **This is the category the next three lessons (L03, L04) go deep on.**

## 🎯 Category 2 — Unsupervised Learning

**The idea in plain English.** You have inputs *but no correct answers*. The model finds structure — groups, patterns, anomalies — in the data on its own.

**Real-world analogy.** A librarian faced with a crate of new books, none labelled by genre. She groups them by *"these feel similar"* — romance here, mystery there — even though no one told her what the genres are.

**A NorthStar example (we will actually do this in L05).** Imagine NorthStar wants to know *what kinds of customers it has* — people who only buy formal wear, people who buy activewear every weekend, people who return everything. Nobody labelled them in advance. An unsupervised algorithm can look at purchase history and propose groupings that Sarah's team can then name and use for targeted marketing.

**Why it matters.** Unsupervised learning is for **exploration** when you have data but no labels. Customer segmentation, anomaly detection in logs, document clustering. **L05 (Unsupervised Learning)** covers this properly.

### 👀 See it happen — 30-second clustering demo

Unsupervised learning is the hardest category to picture, so let's watch it once. The cell below invents 9 customers described by just two numbers (visits per month, average spend) — **with no labels at all** — and asks K-Means to find 3 natural groups. Just run it; you don't need to understand the code yet (Lesson 3.5 is entirely about this).


```python
from sklearn.cluster import KMeans   # K-Means = the classic "find groups in unlabelled data" tool

# 9 customers, each described by [visits per month, average spend in £] — NO labels given
customers = [[2, 15], [3, 18], [1, 12],      # low-visit, low-spend
             [12, 25], [14, 22], [11, 28],   # frequent browsers, modest spend
             [4, 190], [5, 210], [3, 175]]   # rare visits, big spenders

groups = KMeans(n_clusters=3, n_init=10, random_state=42).fit_predict(customers)  # ask for 3 groups

for person, g in zip(customers, groups):                     # show which group each customer landed in
    print(f"visits={person[0]:>2}, spend=£{person[1]:>3}  →  group {g}")
# Notice: nobody told the algorithm what the groups MEAN — it found the structure by itself.
```

## 🎯 Category 3 — Reinforcement Learning

**The idea in plain English.** An agent *takes actions in an environment*, gets *rewards or penalties*, and adjusts its behaviour over time to maximise reward.

**Real-world analogy.** Training a dog with treats. Sit → treat. Bark at the mailman → no treat. Over time the dog learns which actions pay off.

**A NorthStar-ish example.** A product-recommendation system that learns which suggestions *actually get clicked and bought*. Every customer interaction is a reward signal, and the system gets better by trying different suggestions over time.

**Why it matters — and why we will not build one.** Reinforcement learning is behind game-playing AIs, robotic control, large-scale recommendation engines, and parts of how modern LLMs are trained. It is a powerful field, but building an RL system needs an environment the agent can safely try things in (often a simulator) plus millions of trials. For a 10-lesson course on general ML, we just need you to *recognise* the category when you see it in the wild. **Not built in this course.**

## 🧩 Comparison at a glance

| Dimension | Supervised | Unsupervised | Reinforcement |
|---|---|---|---|
| **Data you need** | Features + labels | Features only | Environment + reward signal |
| **Goal** | Predict labels for new data | Find structure | Maximise reward over time |
| **Sarah-world example** | Classify a review positive / negative | Discover hidden customer groups | Recommendation bot that learns from clicks |
| **Covered in this course** | L01, L03, L04, L08, L09, L10 | L05 | Mentioned, not built |

## 💼 Which category dominates business ML — and why?

Rough industry split:

| Category | Share of business ML projects (estimated) |
|---|---|
| Supervised | ~70–80% |
| Unsupervised | ~15–25% |
| Reinforcement | ~1–5% |

**Why supervised dominates:**
- Most business data arrives *with labels* (did the customer churn? did the transaction get reported as fraud? did the user click?).
- Evaluation is straightforward — you can measure accuracy on held-out data.
- The techniques are mature and reliable.

**Why RL is rare in business:**
- Needs an environment the agent can safely interact with — usually a simulator, which is expensive to build.
- Training takes millions of trials.
- Most business decisions can be reframed as supervised if you are willing to collect labels.

**This is why the course spends most of its time on supervised techniques and their deep-learning extensions.** Sarah's sentiment task was the first instance; L03 and L04 go much deeper.

## ✅ Section 2 Summary

| Category | Needs labels? | Goal | When to use |
|---|---|---|---|
| **Supervised** | Yes | Predict labels for new data | Most business ML — churn, fraud, forecasting, sentiment |
| **Unsupervised** | No | Discover structure | Segmentation, anomaly detection, exploratory work |
| **Reinforcement** | No (but needs reward) | Maximise long-term reward | Games, robotics, large-scale recommendations |
